# Experiment 6: Functions

```

# 1.
def find_max_min(seq):
    max_val = seq[0]
    min_val = seq[0]
    for i in seq:
        if i > max_val:
            max_val = i
        if i < min_val:
            min_val = i
    return max_val, min_val

print(find_max_min([10, 6, 8, 90, 12, 56]))
```

```

# 2.
def sum_of_cubes(n):
    s = 0
    for i in range(1, n):
        s += i**3
    return s

print(sum_of_cubes(5))
```

```

# 3.
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n-1)
    print(n)

print_numbers(5)
```

```

# 4.
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

def fibonacci_series(n):
    for i in range(n):
        print(fib(i), end=" ")

fibonacci_series(7)
```

```

# 5.
import math
volume_cone = lambda r, h: (1/3) * math.pi * r * r * h
print(volume_cone(3, 5))
```

```

# 6.
lst = [10, 6, 8, 90, 12, 56]
max_min = lambda l: (max(l), min(l))
print(max_min(lst))
```

```

# 7.
def keyword_arg(name, age):
    print(name, age)

keyword_arg(age=20, name="Rahul")

def default_arg(name, city="Delhi"):
    print(name, city)

default_arg("Aman")

def var_len_arg(*args):
    for i in args:
        print(i)

var_len_arg(1, 2, 3, 4)
```

```

# 8.
d = {'a': 10, 'b': 10, 'c': 10}
check_same = lambda d: len(set(d.values())) == 1
print(check_same(d))
```
